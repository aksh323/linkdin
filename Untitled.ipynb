```python
options = Options()
# options.add_argument('--disable-gpu')
# options.add_argument("--headless")
options.add_argument("--log-level=3")
# options.add_argument('--blink-settings=imagesEnabled=false')
# options.binary_location =r'C:\Users\akash\Desktop\LindIn\driver\chrome.exe'
options.binary_location =r'driver\chrome.exe'

# options.add_argument("--no-sandbox")
driver = webdriver.Chrome(service=Service(executable_path='chromedriver.exe'), options=options)

```

```python
page = driver.page_source
soup = BeautifulSoup(page, 'html.parser')
```

```python
linkdin_Data = {'position':[],'company_name':[],'location':[],'hr_name':[],'hr_position':[],'hr_link':[],'email':[],'phone':[]}
```

```python
data = driver.find_elements(By.CLASS_NAME,'scaffold-layout__list-item')
```

```python

for i in data:
    i.click()
    time.sleep(5)
    data_position = i.text.split('\n')
    linkdin_Data['position'].append(data_position[0])
    linkdin_Data['company_name'].append(data_position[1])
    linkdin_Data['location'].append(data_position[2])
    try:
        hr = driver.find_element(By.CLASS_NAME,'hirer-card__container').text
        hr_ = hr.split('\n')
        linkdin_Data['hr_name'].append(hr_[0])
        linkdin_Data['hr_position'].append(hr_[2])
        hr_link = driver.find_element(By.CLASS_NAME,'hirer-card__hirer-information').find_element(By.CLASS_NAME,'app-aware-link ').get_attribute('href')        
        linkdin_Data['hr_link'].append(hr_link)
    except:
        linkdin_Data['hr_name'].append('nan')
        linkdin_Data['hr_position'].append('nan')
        linkdin_Data['hr_link'].append('nan')
        
    text= driver.find_element(By.CLASS_NAME,'jobs-description-content__text--stretch').text
    try:
        indian_phone_pattern = r'(\+91|91|0)?\s?[6-9]\d{9}'
        indian_phone_numbers = re.search(indian_phone_pattern, text)
        number = indian_phone_numbers.group()
        linkdin_Data['phone'].append(number)
    except:
        linkdin_Data['phone'].append('nan')
        
        
    try:
        email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,7}\b'
        email_addresses = re.findall(email_pattern, text)
        linkdin_Data['email'].append(email_addresses[0])
    except:
        linkdin_Data['email'].append('nan')
        
        
        
        
        
        
    
    
```
